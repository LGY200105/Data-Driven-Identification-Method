```python
# the deep neural network
class DNN(torch.nn.Module):
    def __init__(self, layers):
        super(DNN, self).__init__()
        
        # parameters
        self.depth = len(layers) - 1
        
        # set up layer order dict
        self.activation = torch.nn.Tanh
        
        layer_list = list()
        for i in range(self.depth - 1): 
            layer_list.append(
                ('layer_%d' % i, torch.nn.Linear(layers[i], layers[i+1]))
            )
            layer_list.append(('activation_%d' % i, self.activation()))
            
        layer_list.append(
            ('layer_%d' % (self.depth - 1), torch.nn.Linear(layers[-2], layers[-1]))
        )
        layerDict = OrderedDict(layer_list)
        
        # deploy layers
        self.layers = torch.nn.Sequential(layerDict)
        
    def forward(self, x):
        out = self.layers(x)
        return out
```

```python
class BDNN():
    def __init__(self, input, layers):
        
        self.T = torch.tensor(input[:, 0:1], requires_grad=True).float().to(device)
        self.E = torch.tensor(input[:, 1:2], requires_grad=True).float().to(device)
        self.dE = torch.tensor(input[:, 2:3], requires_grad=True).float().to(device)
        self.d2E = torch.tensor(input[:, 3:4], requires_grad=True).float().to(device)
        self.J = torch.tensor(input[:, 4:5], requires_grad=True).float().to(device)
        self.K = torch.tensor(input[:, 5:6], requires_grad=True).float().to(device)
        self.D = torch.tensor(input[:, 6:7], requires_grad=True).float().to(device)

        self.layers = layers

        self.dnn = DNN(layers).to(device)

        self.optimizer = torch.optim.LBFGS(
            self.dnn.parameters(), 
            lr=0.01, 
            max_iter=50000, 
            max_eval=50000, 
            history_size=500,
            tolerance_grad=1e-5, 
            tolerance_change=1.0 * np.finfo(float).eps,
            line_search_fn="strong_wolfe"       # can be "strong_wolfe"
        )
        

        self.iter = 0

    def net_K(self, a, b, c, d, e):
        K = self.dnn(torch.cat([a, b, c, d, e], dim=1))
        return K
    
    def net_D(self, a, b, c, d, e):
        D = self.dnn(torch.cat([a, b, c, d, e], dim=1))
        return D
    
    def loss_func(self):
        self.optimizer.zero_grad()

        K_pred = self.net_K(self.T, self.E, self.dE, self.d2E, self.J)
        D_pred = self.net_D(self.T, self.E, self.dE, self.d2E, self.J)

        loss_K = torch.mean((self.K - K_pred) ** 2)
        loss_D = torch.mean((self.D - D_pred) ** 2)

        loss = loss_K + loss_D

        loss.backward()
        self.iter += 1
        if self.iter % 10 == 0:
            print(
                'Iter %d, Loss: %.5e, Loss_K: %.5e, Loss_D: %.5e' % (self.iter, loss.item(), loss_K.item(), loss_D.item())
            )
        return loss
    
    def train(self):
        self.dnn.train()

        self.optimizer.step(self.loss_func)

    def predict(self, input_P):
        T = torch.tensor(input_P[:, 0:1], requires_grad=True).float().to(device)
        E = torch.tensor(input_P[:, 1:2], requires_grad=True).float().to(device)
        dE = torch.tensor(input_P[:, 2:3], requires_grad=True).float().to(device)
        d2E = torch.tensor(input_P[:, 3:4], requires_grad=True).float().to(device)
        J = torch.tensor(input_P[:, 4:5], requires_grad=True).float().to(device)
       
        self.dnn.eval()
        K = self.net_K(T, E, dE, d2E, J)
        D = self.net_D(T, E, dE, d2E, J)
        K = K.detach().cpu().numpy()
        D = D.detach().cpu().numpy()
        return K, D
    

```
